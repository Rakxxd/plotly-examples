HISTOGRAM

```
import numpy as np
import random
import pandas as pd
np.random.seed(1234)
```

```
#creating variables
age =np.random.uniform(low= 21 , high = 75 , size = 100)
salary = np.random.normal( loc =3000, scale = 1000, size = 100)
salary_debt = np.random.uniform( low =1, high = 1000, size = 100)
binary_gender = random.choices(['female','male'],k=100)
```

```
#DaraFrame
df = pd.DataFrame({'Age':age,'Salary':salary,'Gender':binary_gender, 'Salary_Debt':salary_debt})
female = df.loc[df.Gender == 'female']
male = df.loc[df.Gender =='male']
```

```
#Histogram
trace = go.Histogram( x=age)
data = [trace]
iplot({'data':data})
```

Histogram is for cotinuous variable as opposed to bar chart

```
#Histogram
trace = go.Histogram( x=age,
                     histnorm = 'probability')
data = [trace]
iplot({'data':data})
```

```
#Overlay
#Histogram
trace = go.Histogram( x=male.Age,
                     name = 'Male Age')
trace1 = go.Histogram( x=female.Age,
                     name='Female Age',opacity=0.5)
data = [trace,trace1]
layout = {'barmode':'overlay'}
iplot({'data':data, 'layout':layout})
```

```
#Stacked
#Histogram
trace = go.Histogram( x=male.Age,
                     name = 'Male Age')
trace1 = go.Histogram( x=female.Age,
                     name='Female Age',opacity=0.5)
data = [trace,trace1]
layout = {'barmode':'stack'}
iplot({'data':data, 'layout':layout})
```

Bin Size

```
#Histogram
trace = go.Histogram( x=age,
                     xbins = dict(start=20, end=80, size=5))
data = [trace]
iplot({'data':data})
```

Cumulative Histogram

```
#Histogram
trace = go.Histogram( x=age,
                     cumulative = dict(enabled=True))
data = [trace]
iplot({'data':data})
```

Changing curve

```
hist_data = [male.Salary.values.tolist(), female.Salary.values.tolist()]
group_labels = ['m','f']

fig = ff.create_distplot(hist_data, group_labels,
                         bin_size = [200,200],
                         colors = ['rgba(255,0,0,1)','rgba(255,255,222,1)'],
                         curve_type ='normal')
#another way to add title
fig.layout.update({'title':'Fitted normal curve'})
iplot(fig)
```

Omit the histogram and rug plot

```
hist_data = [male.Salary.values.tolist(), female.Salary.values.tolist()]
group_labels = ['m','f']

fig = ff.create_distplot(hist_data, group_labels,
                         bin_size = [200,200],
                         curve_type ='normal',
                         show_hist = False,
                         show_rug = False)
#another way to add title
fig.layout.update({'title':'Fitted normal curve',
                  'xaxis': {'title':'Salary', 'range':[1000,5000] }})
iplot(fig)
```

```
np.random.seed(1234)
group_a = np.random.normal(loc=100, scale =10, size =500)
group_b = np.random.normal(loc=110, scale =15, size =500)
control_group = np.random.normal(loc=105, scale =20, size =500)
```

```
#Simple box plot
trace =go.Box(y = group_a)
data = [trace]
iplot({'data':data})
```

```
#More adatsets
trace = []
vals = [group_a,group_b,control_group]
groups = ['Group A','Group B','Group C']
for i in range(0,3):
  trace.append(go.Box(y = vals[i] ,
                      name = groups[i]))
  
data = trace
iplot({'data':data})
```

```
#Outliers
#More adatsets
trace = []
vals = [group_a,group_b,control_group]
groups = ['Group A','Group B','Group C']
for i in range(0,3):
  trace.append(go.Box(y = vals[i] ,
                      name = groups[i],
                      boxpoints='outliers'),
               )
  
data = trace
iplot({'data':data})
```

```
#Outliers
#More adatsets
trace = []
vals = [group_a,group_b,control_group]
groups = ['Group A','Group B','Group C']
for i in range(0,3):
  trace.append(go.Box(y = vals[i] ,
                      name = groups[i],
                      boxpoints='all',
                      jitter = 0.2,
                      pointpos = -1.5),
               )
  
data = trace
iplot({'data':data})
```

```
#Addming a mean
#Outliers
#More adatsets
trace = []
vals = [group_a,group_b,control_group]
groups = ['Group A','Group B','Group C']
for i in range(0,3):
  trace.append(go.Box(y = vals[i] ,
                      name = groups[i],
                      boxpoints='all',
                      jitter = 0.2,
                      pointpos = -1.5,
                      boxmean= True)
               
               )
  
data = trace
iplot({'data':data})
```

```
#Addming a mean
#Outliers
#More adatsets
trace = []
vals = [group_a,group_b,control_group]
groups = ['Group A','Group B','Group C']
for i in range(0,3):
  trace.append(go.Box(y = vals[i] ,
                      name = groups[i],
                      line =dict(color = 'rgb(255,255,255)'),
                      marker = dict(outliercolor = 'rgb(255,100,0)',
                                    symbol = 'circle-open',
                                    size = 10))
               
               )
  
data = trace
iplot({'data':data})
```

```
from plotly import subplots
```

```
trace = go.Scatter (x = female.Age,
                    y=female.Salary,
                    mode = 'markers',
                    marker = dict(size=12, color = female.Salary_Debt, showscale=True,
                    colorscale ='Portland'),
                    )
trace1 = go.Scatter (x = male.Age,
                    y=male.Salary,
                    mode = 'markers',
                    marker = dict(size=12, color = female.Salary_Debt, showscale=True,
                    colorscale ='Portland'),
                    )
fig = subplots.make_subplots(rows=1,
                          cols=2,
                          subplot_titles = ('Female','Male'))
fig.append_trace(trace,1,1)
fig.append_trace(trace1,1,2)
fig.layout.update(title='Correlation between salary and age')

iplot(fig)
```

```
#Shared axis
trace = go.Scatter (x = female.Age,
                    y=female.Salary,
                    mode = 'markers',
                    marker = dict(size=12, color = female.Salary_Debt, showscale=True,
                    colorscale ='Portland'),
                    )
trace1 = go.Scatter (x = male.Age,
                    y=male.Salary,
                    mode = 'markers',
                    marker = dict(size=12, color = female.Salary_Debt, showscale=True,
                    colorscale ='Portland'),
                    )
fig = subplots.make_subplots(rows=1,
                          cols=2,
                          subplot_titles = ('Female','Male'),
                          shared_yaxes = True)
fig.append_trace(trace,1,1)
fig.append_trace(trace1,1,2)
fig.layout.update(title='Correlation between salary and age')

iplot(fig)
```

```
#Shared axis with percentage
trace = go.Scatter (x = female.Age,
                    y=female.Salary,
                    mode = 'markers',
                    marker = dict(size=12, color = 'rgba(255,70,0,0.9)'),
                    )
trace1 = go.Scatter (x = male.Age,
                    y=male.Salary,
                    mode = 'markers',
                    marker = dict(size=12, color = 'rgba(255,0,0,0.9)'),
                     xaxis = 'x2',
                     yaxis = 'y2'
                    )

data = [trace,trace1]
layout = go.Layout(
    title = 'Constrain',
    xaxis=dict(
        domain=[0,0.7]
    ),
    xaxis2=dict(
        domain=[0.8,1]
    ),
    yaxis2=dict(
        anchor='x2'
    )
)

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```
